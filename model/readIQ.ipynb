```python
def read_iq_file(filepath):
    """
    读取 .iq 文件，返回 I/Q 分量
    """
    raw_data = np.fromfile(filepath, dtype=np.float32)
    i_data = raw_data[::2]
    q_data = raw_data[1::2]
    return i_data, q_data
```

```python
# 替换为你的 IQ 文件路径
iq_file_path = "../dataset/raw data/bladerf/bladerf.iq"

# 假设采样率为 20 MHz
sample_rate = 20e6  

# 读取 IQ 数据
i_data, q_data = read_iq_file(iq_file_path)

# 绘图（调用你已有的函数）
plot_iq_data(i_data, q_data, sample_rate, title="真实蓝牙IQ信号可视化")
```
